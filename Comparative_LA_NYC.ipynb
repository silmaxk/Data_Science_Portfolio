With this data source obtained by the Los Angeles Census, it contains information regarding its occupants education levels. Within this dataset, we can sections of the population by their age range, for example "Population 18 to 24 years" and "Population 25 years and over".

<br>

However, for my analysis, I will only be focusing on the information provided for the "Population 18 to 24 years" to keep it fair with my analysis for the education levels in New York City



> Things to keep in mind:

*   We are only focusing on rows 2-5 (excluding row 1 since that is the TOTAL)
*   The numbers listed in the column total_num have commas, so we need to make sure to remove those in order to not receive any errors (they will be considered to be strings and we can't have that when we are trying to plot numbers)





```python
#Here, as mentioned, we are only going to be using the sections "Population 18 to 24 years", so we are only going to be focusing on rows 2-6 and we are saving that as subsets
subset = Education_LA.iloc[2:6]

#We need to make sure the numbers are floats, so we are replacing the commas in the numbers with nothing (removing them) and we are only doing this for the indicated subset portion: rows 2-6
#This step is important to make sure the values are treated as numerical data so we can then graph them
subset['total_num'] = subset['total_num'].str.replace(',', '').astype(float)

#Here we are making sure that we are sotring the values by using the "total_num" column
subset_sorted = subset.sort_values(by='total_num')

#We are then extracting the columns "levels" and "total_num" from the dataframe and saving them
education_levels = subset_sorted['edu_level']
total_population = subset_sorted['total_num']

#This code includes the actual numerical value of the total_num  next to its respective bar
for i, v in enumerate(total_population):
    plt.text(v, i, f'{v:,.0f}', va='center', fontsize=8, color='black')

#To depict this relation, I am plotting a horizontal bar chart
plt.barh(education_levels, total_population, color='pink')

plt.xlabel('Total Population')
plt.ylabel('Education Level')
plt.title('Education Levels for Age Group 18-24 in Los Angeles')

plt.show()
```

From this graph of the Education levels in Los Angeles, we can see that most people have some college or associate's degree level while a few people have completely finished their bachelor's degree or went for higher education.

For the New York City Education levels at first glance is similar to the Education Levels for Los Angeles in regards to the highest population having some college or associate's degree and the least being less than high school graduate

I will now create a stacked horizontal bar chart to visually compare and contrast the similarities or differences between the Education Levels in Los Angeles and New York City

```python
# Extract the relevant subset of rows for both cities
subset_NYC = Education_NYC.iloc[2:6]
subset_LA = Education_LA.iloc[2:6]

# Convert total population to float for both cities
subset_NYC['total_num'] = subset_NYC['total_num'].str.replace(',', '').astype(float)
subset_LA['total_num'] = subset_LA['total_num'].str.replace(',', '').astype(float)

# Sort the subsets by total population in ascending order for both cities
subset_sorted_NYC = subset_NYC.sort_values(by='total_num')
subset_sorted_LA = subset_LA.sort_values(by='total_num')

# Extract education levels and total population from the sorted subsets for both cities
education_levels_NYC = subset_sorted_NYC['edu_level']
total_population_NYC = subset_sorted_NYC['total_num']

education_levels_LA = subset_sorted_LA['edu_level']
total_population_LA = subset_sorted_LA['total_num']

fig, ax = plt.subplots()

# Set the width of the bars
bar_width = 0.35

# Create grouped horizontal bar graph for NYC and LA
bar_positions_NYC = np.arange(len(education_levels_NYC))
bar_positions_LA = np.arange(len(education_levels_LA)) + bar_width

# Plot bars for NYC
ax.barh(bar_positions_NYC, total_population_NYC, color='purple', label='NYC', height=bar_width)

# Add labels for NYC
for i, v in zip(bar_positions_NYC, total_population_NYC):
    ax.text(v, i, f'{v:,.0f}', va='center', fontsize=8, color='black')

# Plot bars for LA
ax.barh(bar_positions_LA, total_population_LA, color='pink', label='LA', height=bar_width)

# Add labels for LA
for i, v in zip(bar_positions_LA, total_population_LA):
    ax.text(v, i, f'{v:,.0f}', va='center', fontsize=8, color='black')

ax.set_xlabel('Total Population')
ax.set_ylabel('Education Level')
ax.set_title('Education Levels for Age Group 18-24 in NYC and LA')

ax.set_yticks(bar_positions_NYC + bar_width / 2)
ax.set_yticklabels(education_levels_NYC)

ax.legend()

plt.show()
```

Looking at the statistics of the graph, it seems like more people in Los Angeles, more than 100,000 more people have some college or associate's degree compared to those in New York City. While Los Angeles seems to have greater statistics than New York City, there are more people in New York City that have a Bachelor's degree or higher.

Taking a look at the information presented to us through the Census, we are given percentage values of the amount of people that are within the stated income bracket.

<br>

I will be using a pie chart to visualize the the break down of the income brackets and the percentage of how the overall population that is part of that bracket

<br>

Note:


*   Although in the dataset they are shown as percentages, they are still strings, we need to remove the commas and % from the column and convert them to floats
*   There are multiple seconds in this dataset, however, I will only be focusing on the most recent data (2022) and use the information only for the income brackets, not the total or mean and median, so I will only be needed rows 1-10



```python
#Only focusing on the income brackets, so rows 1-10
subset = Income_LA.iloc[1:10]

# Extract income levels and total percentages from the subset
income_levels = subset['income_level']

# Convert '##.#%' strings to numerical values
total_percentages = subset['total_percentage'].replace({',': '', '%': ''}, regex=True).astype('float') / 100.0

# Plotting
fig, ax = plt.subplots()

# Create a pie chart with actual percentage values
ax.pie(total_percentages, labels=income_levels, autopct='%1.1f%%', startangle=90)

# Add a title
ax.set_title('Income Distribution in LA')

# Show the plot
plt.show()
```

I will also include labels for those markers so it will state the type of offense that was reported in this area.

Through this analysis, we will be able to see the areas with the most crime, and in general see the crime distribution across the cities.



> Crime_LA



```python
Crime_LA.head()
```

As stated, we will be using the Lat and Lon (Latitude & Longitude) values given in the dataset to create markets of where these crimes took place and use the offense column to create a popup text for the respective Lat and Lon values

```python
#I do not want to do all of the information presented in the dataset since it is too many entries
Sub_Crime_LA = Crime_LA.head(100)

# Create a Folium map centered at the mean latitude and longitude listed in the dataset
map_center = [Sub_Crime_LA['Lat'].mean(), Sub_Crime_LA['Lon'].mean()]
crime_la_map = folium.Map(location=map_center, zoom_start=12)

# Add markers to the map (the offenses)
for index, row in Sub_Crime_LA.iterrows():
    popup_text = f"Offense: {row['offense']}"
    folium.Marker(location=[row['Lat'], row['Lon']], popup=popup_text).add_to(crime_la_map)

# Save the map to an HTML file as backup
crime_la_map.save("crime_la_map.html")

crime_la_map
```

From this geo-analysis, we can see that most of the 100 entries are located around the heart of Los Angeles. While there are still other crimes being reported from around, in Los Angeles it seems like there is one area that causes more issues than the rest.
