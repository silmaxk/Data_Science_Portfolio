# *Does Employment status directly correlate with the mental health?*





```python
#Import all libraries needed to show relation between both data sources

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import pie, axis, show
%matplotlib inline
```

```python
#Check to see if both datasets are good to be used
#First check the HEALTH dataset

health = pd.read_csv('health.csv')

health.head()
```

```python
#Now check the POVERTY dataset

poverty = pd.read_csv('poverty.csv')

poverty.head()
```

```python
#UNEMPLOYED
#TOTAL unemployed people
unemployed_total_row = 35
unemployed_total_column = 'United States!!Total!!Estimate'
unemployed_total = poverty.loc[unemployed_total_row, unemployed_total_column]

print("The TOTAL Unemployment is: " + unemployed_total)

#MALE unemployed people
unemployed_male_row = 36
unemployed_male_column = 'United States!!Total!!Estimate'
unemployed_male = poverty.loc[unemployed_male_row, unemployed_male_column]

print("The total MALE Unemployment is: " + unemployed_male)

#FEMALE unemployed people
unemployed_female_row = 37
unemployed_female_column = 'United States!!Total!!Estimate'
unemployed_female = poverty.loc[unemployed_female_row, unemployed_female_column]

print("The total FEMALE Unemployment is: " + unemployed_female)

```

```python
#EMPLOYED
#TOTAL employed
employed_total_row = 32
employed_total_column = 'United States!!Total!!Estimate'
employed_total = poverty.loc[employed_total_row, employed_total_column]

print("The TOTAL employment is: " + employed_total)


#MALE employed
employed_male_row = 33
employed_male_column = 'United States!!Total!!Estimate'
employed_male = poverty.loc[employed_male_row, employed_male_column]

print("The total MALE employed is: " + employed_male)


#FEMALE employed
employed_female_row = 34
employed_female_column = 'United States!!Total!!Estimate'
employed_female = poverty.loc[employed_female_row, employed_female_column]

print("The total FEMALE employed is: " + employed_female)
```

```python
#Let's take a look at the numbers a bit closer

#UNEMPLOYED
print("The TOTAL Unemployment is: " + unemployed_total)
print("The total MALE Unemployment is: " + unemployed_male)
print("The total FEMALE Unemployment is: " + unemployed_female)

print()

#EMPLOYED
print("The TOTAL employment is: " + employed_total)
print("The total MALE employed is: " + employed_male)
print("The total FEMALE employed is: " + employed_female)

```

```python
plt.figure(1)
sns.countplot(data=health, x='have_depression')
plt.title("Mental Health")
plt.xlabel('Depression')
plt.ylabel("Count")
plt.show()
```

Looking at the difference between those unemployed and employed as well as the statistics of people who are suffering from depression, we can see that employment rates is not heavily affected by mental health
- Out of approximately 16,000 people, half of them suffer from depression
- There is a total of 155,426,604 people that are employed and 10,400,638 people that are unemployed (not out of the employed people)
- There is a large jump from those who are unemployed to employed
- The half statistics of those suffering from mental health issues is not statistically relevant enough to conclude that the unemployment rates are directly affected by mental health issues
